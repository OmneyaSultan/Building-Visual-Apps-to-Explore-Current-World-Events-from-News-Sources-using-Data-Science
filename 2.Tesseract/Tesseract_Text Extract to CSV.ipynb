```python
import pytesseract
import shutil
import os
import random
try:
 from PIL import Image
except ImportError:
 import Image
import pandas as pd
```

```python
inputPath="/content/drive/MyDrive/HW3/HW3/input/fox_cropped"
for imageName in os.listdir(inputPath):
              
        imagePath = os.path.join(inputPath, imageName)
        
  
        # applying ocr using pytesseract for python
        extract = pytesseract.image_to_string(Image.open(imagePath))
        extract=''.join(extract)
        extract=extract.split('\n')
        lst=list()
        for i in extract:
          if i!='' and i!=' ':lst.append(i)

        imageName=imageName.replace('.png','.csv')
        

        os.chdir('/content/drive/MyDrive/HW3/HW3/output/fox_cropped')
      
    
        df = pd.DataFrame(lst) 
        df.to_csv(imageName) 

        os.chdir('/content/drive/MyDrive/HW3/HW3/input/fox_cropped')
```
