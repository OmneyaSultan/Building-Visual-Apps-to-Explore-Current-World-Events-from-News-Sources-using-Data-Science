**World Map**

```python
#define animation function 
def animate_map(time_col):
    fig = px.scatter_mapbox(df, #use data in df with plotly express and mapbox 
              lat="Latitude" , #use latitude from df
              lon="Longitude", #use longitude from df
              hover_name="Country", #use country as hover name from df
              #color="Country",
              animation_frame=time_col, #use passed date argument from df as animation frame
              mapbox_style='carto-positron', #set animation style
              category_orders={ #define unique categories from date data
              time_col:list(np.sort(df[time_col].unique()))
              },                  
              zoom=1 #set zoom value
              )
    fig.update_traces(marker=dict(size=12)) #set marker size 
    fig.update_layout(showlegend=False) #remove legend
    fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1000 #set animation speed
    fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 1000 #set animation speed
    fig.layout.sliders[0].pad.t = 10
    fig.show();
animate_map(time_col='Date')
```

```python
def animate_map2(time_col):
    fig = px.scatter_mapbox(df2,
              lat="Latitude" ,
              lon="Longitude",
              hover_name="GPE",
              #color="GPE",
              mapbox_style='carto-positron',
              animation_frame=time_col,
              category_orders={
              time_col:list(np.sort(df2[time_col].unique()))
              },                  
              zoom=5
              )
    fig.update_mapboxes(center_lon=31.2718321)
    fig.update_mapboxes(center_lat=49.4871968)
    fig.update_layout(showlegend=False)
    fig.update_traces(marker=dict(size=12))
    fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1000
    fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 1000
    fig.layout.sliders[0].pad.t = 10
    fig.show();
animate_map2(time_col='Date')
```
